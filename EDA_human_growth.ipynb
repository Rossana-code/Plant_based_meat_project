<span style="font-size:3em;"> Analysis of the global plant-based meat market </span> 

This is an analysis of the global plant-based meat market, that take into consideration the following three axes for each country: water shortages, the meat consumption in relation to the production and the percentage of vegetarians, creating a coefficient to predict which country will be more suited to the market. 

The following notebook is the cleaning and prep of the population growth part. 
This CSV file, along to the other parts of the analysis, has been used to build a dashbord in Tableau, that you can find here:

```python
# Rename the remaining columns.
dataset_pop=dataset_pop.rename(columns={"ISO3 Alpha-code": "Location_code","Region, subregion, country or area *":"Regions","Total Population, as of 1 July (thousands)":"Total Population"})
dataset_pop.head()
```

```python
dataset_pop["Total Population"] = pd.to_numeric(dataset_pop["Total Population"], errors='coerce')
dataset_pop['Total Population'].unique()
```

```python
dataset_pop[["Total Population"]] = dataset_pop[["Total Population"]].astype(int).round()
```

```python
dataset_pop["Location_code"].unique()
```

```python
# change Nan in null to be able to run 'is not null'
dataset_pop["Location_code"].fillna("null")
```

```python
# divide countries from economic zones and continents
dataset_pop["Zone"] = dataset_pop.apply(lambda x: x["Regions"] if pd.isnull(x["Location_code"]) else "Not a Zone", axis=1)
```

```python
# Check economic zones
dataset_pop["Zone"].unique()
```

```python
# divide countries from economic zones and continents
dataset_pop["Country"] = dataset_pop.apply(lambda x: x["Regions"] if pd.notnull(x["Location_code"]) else "Nan", axis=1)
```

```python
# Check countries
dataset_pop['Country'].unique()
```

```python
display(dataset_pop)
```

```python
# Create a column Continent and add continents only.

dataset_pop['Continent'] = dataset_pop['Zone'].apply(lambda x:
                                                   'South America' if x == 'South America'
                                                   else 'North America' if x == 'NORTHERN AMERICA'
                                                   else 'Africa' if x =='AFRICA'
                                                   else 'Asia' if x =='ASIA'
                                                   else 'Europe' if x =='EUROPE'
                                                   else 'Oceania' if x =='OCEANIA'
                                            else "Not a continent")
display(dataset_pop)
```

```python
dataset_pop = dataset_pop.drop(columns=["Regions"])
```

```python
display(dataset_pop)
```

```python
dataset_pop['Year'] = dataset_pop['Year'].fillna(0)
```

```python
dataset_pop["Year"] = dataset_pop["Year"].fillna(pd.NA).astype(pd.Int64Dtype())
```
